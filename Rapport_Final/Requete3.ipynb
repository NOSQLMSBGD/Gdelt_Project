# Articles positives and negatives

```python
pos_Ac1 = pd.read_csv('results_pos_Act1.csv')
pos_Ac2 = pd.read_csv('results_pos_Act2.csv')
```

```python
pos_Ac1 = pd.read_csv('results_pos_Act1.csv')
pos_Ac2 = pd.read_csv('results_pos_Act2.csv')

pos_Ac1['sum_pos'] = pos_Ac1['f0_']
pos_Ac2['sum_pos'] = pos_Ac2['f0_']
pos_Ac2 = pos_Ac2.set_index('Actor2Code')
pos_Ac1 = pos_Ac1.set_index('Actor1Code')
pos_Ac1.head()
pos = pd.merge(pos_Ac1,pos_Ac2,left_index=True, right_index=True)
pos.head()
pos['pos']=pos['sum_pos_x']+pos['sum_pos_y']
list_=['f0__x', 'f0__y','sum_pos_x','sum_pos_y']
pos = pos.drop(list_, axis=1)
```

```python
neg_Ac1 = pd.read_csv('neg_Ac1.csv')
neg_Ac2 = pd.read_csv('neg_Ac2.csv')
neg_Ac1['sum_neg'] = neg_Ac1['f0_']
neg_Ac2['sum_neg'] = neg_Ac2['f0_']
neg_Ac2 = neg_Ac2.set_index('Actor2Code')
neg_Ac1 = neg_Ac1.set_index('Actor1Code')
neg_Ac1.head()
neg = pd.merge(neg_Ac1,neg_Ac2,left_index=True, right_index=True)
neg.head()
neg['neg'] = neg['sum_neg_x']+neg['sum_neg_y']
list_=['f0__x', 'f0__y','sum_neg_x','sum_neg_y']
neg = neg.drop(list_, axis=1)
```

```python
article = pd.merge(neg,pos,left_index=True, right_index=True)
article['total']=article['pos']+article['neg']
article['prop_plus']=article['pos']/article['total']
article['prop_moins']=article['neg']/article['total']
article.sort_values(by='prop_plus')
```

# Remove non country

```python
country = pd.read_csv('CAMEO.country.csv',sep='\t')
codelist = country['CODE'].values
codet = codelist.transpose()
codet = pd.DataFrame({'Country':codet}).set_index('Country')
codet.head()
```

```python
article_clean = pd.merge(article,codet,left_index=True, right_index=True)
p_plus = article_clean.sort_values(by='prop_plus', ascending=False)
p_moins =  article_clean.sort_values(by='prop_moins', ascending=False)
```

# Map

## Map 1 

Carte qui visualise les 2 pays ayant reçues le plus d'avis positifs (en vert) et d'avis negatifs (en rouge)

```python
bcr_rgba = [(0.02, 0.2, 0.4, 1),    # grayish blue, opaque
            (0.02, 0.2, 0.4, 0.3),  # grayish blue, transparent
            (0.4,  0.0, 0.1, 0.3),  # dark red, transparent
            (0.4,  0.0, 0.1, 1)]    # dark red, opaque
import random
import pandas as pd

import folium

from branca.colormap import LinearColormap

import json



map_data = pd.DataFrame({

    'A3':[p_plus.index[0],p_moins.index[0]],

    'value':[p_plus.iloc[0,3]*10,p_moins.iloc[0,4]*10]

})



map_dict = map_data.set_index('A3')['value'].to_dict()



color_scale = LinearColormap(['green','red'], vmin = min(map_dict.values()), vmax = max(map_dict.values()))
color_scale = LinearColormap(bcr_rgba, vmin = min(map_dict.values()), vmax = max(map_dict.values()))


def get_color(feature):

    value = map_dict.get(feature['properties']['A3'])

    if value is None:

        return '#dfdfdf' # MISSING -> gray

    else:

        return color_scale(value)



m = folium.Map(

    location = [35, 5], 

    zoom_start = 2,

    tiles='Open Street Map', #Open Street Map #Stamen Terrain #

    control_scale=True,

    no_wrap=True,

    width=950, height=500

)



folium.GeoJson(

    data = 'https://github.com/simonepri/geo-maps/releases/download/v0.6.0/countries-land-10km.geo.json',

    style_function = lambda feature: {

        'fillColor': get_color(feature),

        'fillOpacity': 0.7,

        'color' : 'black',

        'weight' : 0.2,

    }    

).add_to(m)
```
